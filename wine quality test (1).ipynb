```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
class Wine:
    def checkqualityred(self,dataset):
        self.dataset=dataset
        self.dataset.plot(x='quality', y='fixed acidity', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='volatile acidity', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='citric acid', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='residual sugar', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='chlorides', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='free sulfur dioxide', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='total sulfur dioxide', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='density', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='pH', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='sulphates', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='alcohol', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.hist(figsize=(10, 10))
        plt.show()
        array = self.dataset.values
        x = array[:, 0:11]
        y = array[:, 11]
        validation_size = 0.1
        seed = 60
        x_train, x_validation, y_train, y_validation = model_selection.train_test_split(x, y, test_size=validation_size,random_state=seed)
        seed = 60
        lda = LinearDiscriminantAnalysis()
        lda.fit(x_train, y_train)
        predictions = lda.predict(x_validation)
        print("CONFUSION_MATRIX:\n", confusion_matrix(y_validation, predictions))
        print("CLASSIFICATION REPORT=\n", classification_report(y_validation, predictions))
        print("ACCURACY=\n", accuracy_score(y_validation, predictions))
    def checkqualitywhite(self,dataset):
        self.dataset = dataset
        self.dataset.plot(x='quality', y='fixed acidity', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='volatile acidity', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='citric acid', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='residual sugar', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='chlorides', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='free sulfur dioxide', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='total sulfur dioxide', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='density', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='pH', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='sulphates', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.plot(x='quality', y='alcohol', kind="scatter", layout=(6, 4), figsize=(10, 4))
        plt.show()
        self.dataset.hist(figsize=(10, 10))
        plt.show()
        array = self.dataset.values
        x = array[:, 0:11]
        y = array[:, 11]
        validation_size = 0.1
        seed = 60
        x_train, x_validation, y_train, y_validation = model_selection.train_test_split(x, y, test_size=validation_size,
                                                                                        random_state=seed)
        seed = 60
        lda = DecisionTreeClassifier()
        lda.fit(x_train, y_train)
        predictions = lda.predict(x_validation)
        print("CONFUSION_MATRIX:\n", confusion_matrix(y_validation, predictions))
        print("CLASSIFICATION REPORT=\n", classification_report(y_validation, predictions))
        print("ACCURACY=\n", accuracy_score(y_validation, predictions))

w=Wine()
print("*****  üç∑ üç∂ üç∑ üç∂ üç∑--Welcome to Wine Quality Checking System--üç∑ üç∂ üç∑ üç∂ üç∑  *****")
print("\nChoose  Any One Type of WINE")
print("1. WHITE WINE üç∂\n2. RED WINE üç∑\n")
ch = int(input("Enter Your Choice:\n"))
if ch == 1:
    f = "winequality-white.csv"
    dataset = pd.read_csv("C:\\Users\\sayan\\Downloads\\{}".format(f))
    dataset.info()
    w.checkqualitywhite(dataset)
else:
    g = "winequality-red.csv"
    dataset1 = pd.read_csv("C:\\Users\\sayan\\Downloads\\{}".format(g))
    dataset1.info()
    w.checkqualityred(dataset1)
```
